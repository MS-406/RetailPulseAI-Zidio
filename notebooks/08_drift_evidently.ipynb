# Day 12: Drift Detection with Evidently AI\nThis notebook sets up a formal drift detection pipeline using `evidently`. It compares older 'reference' data with newer 'current' data to detect statistical drift.

```
import pandas as pd
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset, TargetDriftPreset
import warnings

warnings.filterwarnings('ignore')

# 1. Load the cleaned dataset
df = pd.read_csv('../data/cleaned/cleaned_retail.csv')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Define features for drift detection
features = ['Quantity', 'UnitPrice']
target = 'TotalPrice'

df = df[features + [target, 'InvoiceDate']].dropna()

# 2. Split into Reference (first 6 months) and Current (last 6 months)
midpoint = df['InvoiceDate'].min() + (df['InvoiceDate'].max() - df['InvoiceDate'].min()) / 2
reference = df[df['InvoiceDate'] <= midpoint]
current = df[df['InvoiceDate'] > midpoint]

print(f"Reference shape: {reference.shape}")
print(f"Current shape: {current.shape}")
```

```
# 3. Create Evidently Report
report = Report(metrics=[
    DataDriftPreset(),
    TargetDriftPreset()
])

report.run(reference_data=reference, current_data=current, column_mapping=None)

# 4. Save report as HTML
report.save_html('../reports/week2_evidently_drift_report.html')
print("Evidently AI Drift Report saved to '../reports/week2_evidently_drift_report.html'")
```
